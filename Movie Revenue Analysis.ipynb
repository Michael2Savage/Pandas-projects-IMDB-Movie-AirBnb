```python
df = pd.read_csv('imdb-movies.csv')
```

### Drop columns without neccesary information and remove all records with no financial information -- Pay close attention to things that don't tell you anything regarding financial data

```python
data = df.drop(['id', 'imdb_id', 'popularity', 'cast', 'homepage', 'tagline', 'keywords', 'overview', 'vote_count', 'vote_average', 'release_year', 'runtime'], inplace = False, axis = 1)
data
```

### Data Cleaning

```python
def seperate(input1):
    try:
        newlist = input1.split('|')
        return newlist
    except:
        return 'N/A'

data['production_companies'] = data['production_companies'].apply(seperate)
prod = data.explode('production_companies', ignore_index=False)
prod

```

## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
# dates = prod[['release_date']]
# for release_date in prod:
#     if 'release_date'[-1] == 5:
#         print('release_date')


prod.groupby(['production_companies']).count().sort_values('original_title', ascending=False).head(6)[['original_title']]

# The top 5 production companies are Universal, WB, Paramount, 20th century fox, and columbia pictures.
```

### What 5 movie genres grossed the highest all-time?

```python
data['genres'] = data['genres'].apply(seperate)
data_genres = data.explode('genres', ignore_index=False)


```

```python
data_genres.groupby('genres').sum()[['revenue']].sort_values('revenue', ascending = False).head()

# The top 5 genres in terms of revenue is action, adventure, comedy, drama, and thriller.
```

### Who are the top 5 grossing directors?

```python
data.groupby('director').sum()[['revenue']].sort_values('revenue', ascending = False).head()

# the highest grossing directors are Steven Spielberg, Peter Jackson, James Cameron, Michael Bay, Christopher Nolan.
```
